```python
'''
Plot the magnitude and phase of each kernel/image, employing the fftshift() command to rearrange
frequencies. Also using surf() command, plot the 3D magnitude of the Fourier transform
'''

import numpy as np
import matplotlib.pyplot as plt
from PIL import Image
# Open the image file
image_1 = Image.open("Ex1_1.png")
image_1 = image_1.convert("L")
image_2 = Image.open("Ex1_2.png")
image_2 = image_2.convert("L")
image_3 = Image.open("Ex1_3.png")
image_3 = image_3.convert("L")
image_4 = Image.open("Ex1_4.png")
image_4 = image_4.convert("L")

def create_fig(image):
    fig = plt.subplots(1, 5, figsize=(18, 5))

    ax1 = plt.subplot(151)
    ax1.title.set_text('Original Image')
    ax1.imshow(image, cmap='gray')

    ax2 = plt.subplot(152)
    ft = np.fft.fft2(image)
    ft = np.log10(np.fft.fftshift(ft))
    ax2.title.set_text('2D Fourier Transform')
    ax2.imshow(np.abs(ft))

    ax3 = plt.subplot(153, projection='3d')
    X, Y = np.mgrid[0:ft.shape[0], 0:ft.shape[1]]
    ax3.title.set_text('3D Fourier Transform')
    ax3.plot_surface(X, Y, np.abs(ft), cmap='viridis')

    ax4 = plt.subplot(154)
    ax4.title.set_text('Phase')
    ax4.imshow(np.angle(ft), cmap='gray')

    ax5 = plt.subplot(155)
    inv_ft = np.fft.ifftshift(10**ft)
    inv_ft = np.fft.ifft2(inv_ft)
    ax5.title.set_text('Reconstructed Image')
    ax5.imshow(abs(inv_ft), cmap='gray')
    
    plt.show()

create_fig(image_1)
create_fig(image_2)
create_fig(image_3)
create_fig(image_4)

plt.savefig('Ex2_1_3D.png')
plt.show()
```
