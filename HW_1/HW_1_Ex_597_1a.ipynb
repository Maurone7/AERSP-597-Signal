```python
import numpy as np
import matplotlib.pyplot as plt

u = lambda t: 1 if t >= 0 else 0

f = lambda t: 1/3*t if t<=3 and t>=0 else 0
f_matrix = [f(t) for t in np.linspace(-4, 4, 1000)]
h = lambda t: 1 if t>=-1 and t<=1 else 0
h_matrix = [h(t) for t in np.linspace(-4, 4, 1000)]
plt.plot(np.linspace(-4, 4, 1000), f_matrix), plt.plot(np.linspace(-4, 4, 1000), h_matrix), plt.show()

x = np.linspace(-8, 8, 1999)
            
plt.plot(np.linspace(-4, 4, 1000), f_matrix, label='f(t)')
plt.plot(np.linspace(-4, 4, 1000), h_matrix, label='h(t)')

f = lambda t: -1/3 *t if t+3>=-1 and t+3<=2 else 0
f_matrix = [f(t) for t in np.linspace(-4, 4, 1000)]

plt.fill_between(np.linspace(-4, 4, 1000), [0]*1000, f_matrix, where=(np.linspace(-4, 4, 1000) >= -1) & (np.linspace(-4, 4, 1000) <= 0), color='gray', alpha=0.5)
plt.legend(), plt.savefig('HW_1_Ex_597_1a_1_png'), plt.show()

f = lambda t: -1/3 *t if t+3>=0 and t<=0 else 0
f_matrix = [f(t-2) for t in np.linspace(-4, 4, 1000)]
plt.plot(np.linspace(-4, 4, 1000), f_matrix, label='f(t)')
plt.plot(np.linspace(-4, 4, 1000), h_matrix, label='h(t)')

plt.fill_between(np.linspace(-4, 4, 1000), [0]*1000, f_matrix, where=(np.linspace(-4, 4, 1000) >= -1) & (np.linspace(-4, 4, 1000) <= 1), color='gray', alpha=0.5)
plt.legend(), plt.savefig('HW_1_Ex_597_1a_2_png'), plt.show()
```
